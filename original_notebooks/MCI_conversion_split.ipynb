```python
convlist=list(conv_with_time.index)
```

```python
mciconv=[]
mcinoconv=[]
totalconv=0
totalnoconv=0
for i in mcies:
    idy=i[:10]
    if idy in convlist:
        totalconv=totalconv+1
        if idy in mciconv:
            print("Already added")
        else:
            print("Adding to conv: ", idy)
            mciconv.append(idy)
        
    else:
        totalnoconv=totalnoconv+1
        if idy in mcinoconv:
            print("Already added")
        else:
            print("Adding to noconv: ", idy)
            mcinoconv.append(idy)
```

```python
print("Amount of converted MCIs: ", len(mciconv))
print("Number of converted MCIs studies: ", totalconv)
print("Mean amount of studies per converted MCI: ", totalconv/len(mciconv))
print("Amount of stable MCIs: ", len(mcinoconv))
print("Number of non converted MCIs studies: ", totalnoconv)
print("Mean amount of studies per non converted MCI: ", totalnoconv/len(mcinoconv))
print("Total MCIs: ", len(mciconv)+len(mcinoconv))
print("Total number of studies: ", totalnoconv+totalconv)
```
